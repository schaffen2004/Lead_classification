```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_path = "/home/schaffen/Workspace/Project/Lead_classification/data/data/raw/demo.csv"
```

```python

df = pd.read_csv(raw_path)
df.head()
```

```python
df_processed = df.copy()
```

```python

categorical_features  = ['size', 'role', 'channel']
for col in categorical_features :
    print(f"{col}: {df_processed[col].unique()}")
```

```python
df_encoded = pd.get_dummies(df_processed, columns=categorical_features , drop_first=True)
df_encoded
```

```python
mapping = {
    "Cold": 1,
    "Warm": 2,
    "Hot": 3
}

df_encoded["target"] = df_encoded["target"].map(mapping)
df_encoded
```

```python

import os

out_dir = '/home/schaffen/Workspace/Project/Lead_classification/data/data/processed'
os.makedirs(out_dir, exist_ok=True)
out_path = os.path.join(out_dir, 'df_processed.csv')

df_encoded.to_csv(out_path, index=False)
print(f"Exported df_processed to {out_path}")
```
